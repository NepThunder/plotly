```python
import pandas as pd
import plotly.express as px

file_path = "metadata.csv"  
df = pd.read_csv(file_path)

impedance_data = df[df["type"] == "impedance"].copy()

impedance_data=impedance_data[~impedance_data['start_time'].str.contains(r'\+', na=False)]

impedance_data["aging_metric"] = range(1, len(impedance_data) + 1)

impedance_data_re_sorted = impedance_data.sort_values(by="Re", ascending=True)

# Plot Estimated Electrolyte Resistance (Re) over aging
fig_re = px.scatter(
    impedance_data_re_sorted,
    x="aging_metric",
    y="Re",
    title="Change in Electrolyte Resistance (Re) Over Battery Aging",
    labels={"aging_metric": "Battery Aging (Cumulative Test Number)", "Re": "Electrolyte Resistance (Ohms)"}
)
fig_re.update_traces(marker=dict(size=6, color='blue'))
fig_re.show() 

impedance_data_rct_sorted = impedance_data.sort_values(by="Rct", ascending=True)

# Plot Estimated Charge Transfer Resistance (Rct) over aging
fig_rct = px.scatter(
    impedance_data_rct_sorted,
    x="aging_metric",
    y="Rct",
    title="Change in Charge Transfer Resistance (Rct) Over Battery Aging",
    labels={"aging_metric": "Battery Aging (Cumulative Test Number)", "Rct": "Charge Transfer Resistance (Ohms)"}
)
fig_rct.update_traces(marker=dict(size=6, color='green'))
fig_rct.show()  
```
